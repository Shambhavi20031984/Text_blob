```python
import pandas as pd
```

```python
df = pd.read_csv("News_Headlines.csv")

headlines = df["Headlines"]

from textblob import TextBlob

def sentiment_analyzer1(headlines):

  sentiments1 = []
  for headline in headlines:
       sentiment1 = TextBlob(headline)
       score = sentiment1.sentiment
       sentiments1.append(score)
  return sentiments1


sentiments1 = sentiment_analyzer1(headlines)
# # #print(sentiments)

df["TextBlob Score"] = sentiments1
df.to_csv("News_Headlines.csv", index=False)
```
